***Create various type of plots/charts like histograms, plot based on sine/consine function based on data from a matrix. Further label different axes in a plot and data in a plot.***

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for plotting
x = np.linspace(0, 10, 100)  # Generate 100 evenly spaced points from 0 to 10
y_sin = np.sin(x)  # Sine function values
y_cos = np.cos(x)  # Cosine function values

# Create a histogram
data = np.random.randn(1000)  # Generate random data
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, alpha=0.5, color='blue', edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Create a plot based on sine and cosine functions
plt.figure(figsize=(8, 6))
plt.plot(x, y_sin, label='Sin', color='blue')
plt.plot(x, y_cos, label='Cos', color='red', linestyle='dashed')
plt.title('Sine and Cosine Functions')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
